```python
import selenium
from selenium import webdriver
import pandas as pd
from bs4 import BeautifulSoup
import requests
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
import time
```

```python
from selenium.webdriver.common.by import By
```

```python

from selenium.webdriver.chrome.options import Options
```

```python
driverPATH =  r'/home/alby/Downloads/chromedriver'
browserBIN = '/usr/bin/vivaldi'
```

```python
options = webdriver.ChromeOptions()
options.binary_location = browserBIN
```

```python
driver = webdriver.Chrome(driverPATH, options=options)
```

```python
soup=[]
ur="https://www.malayalachalachithram.com/listsongs.php?tot=1189&g=15&p="
for pno in range(1,41):
    driver.get(ur+str(pno))
    time.sleep(4)
    element = driver.find_elements(By.TAG_NAME, "td")
    
    for i in element:
        soup.append(i.text)
```

```python
slno=soup[0::7]
song=soup[1::7]
movie=soup[2::7]
year=soup[3::7]
singers=soup[4::7]
lyricist=soup[5::7]
composer=soup[6::7]
```

```python
len(singers)
```

```python
Jayachandran_songs=pd.DataFrame({})
Jayachandran_songs["no"]=slno
Jayachandran_songs["song"]=song
Jayachandran_songs["movie"]=movie
Jayachandran_songs["year"]=year
Jayachandran_songs["singers"]=singers
Jayachandran_songs["lyricist"]=lyricist
Jayachandran_songs["composer"]=composer
```

```python
Jayachandran_songs.to_csv('Pjayachandran_disco_mal.csv')
```
